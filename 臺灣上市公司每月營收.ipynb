```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
# For extracting "even" dataframe 
even = np.arange(2, 59, 2)
odd = np.arange(1, 56, 2)

result = pd.DataFrame()
```

```python
# creating the columns' name 
columnsname = []
urlname = []
for i in np.arange(96, 112, 1): 
    for j in np.arange(1, 13, 1): 
        ym = str(i)+'_'+str(j)
        columnsname.append(ym)
        
for r in columnsname[:-120]: 
    url = 'https://mops.twse.com.tw/nas/t21/sii/t21sc03_'+str(r)+'.html'
    urlname.append(url)
    
for h in columnsname[-120:]:
    urll = 'https://mops.twse.com.tw/nas/t21/sii/t21sc03_'+str(h)+'_0.html'
    urlname.append(urll)
```

```python
# First, creating industry list first 
final_df = pd.DataFrame()

url = urlname[0]   
res = requests.get(url) 
chineese_encoding = res.apparent_encoding
res.encoding = chineese_encoding
soup = BeautifulSoup(res.text,"html.parser") 

industryname = []
for jjj in odd:
    data = soup.select('table')[jjj]
    df = pd.read_html(data.prettify())
    dfs = df[0]
    industryname.append(dfs.columns[0][4:])
```

```python
## 規則
## 96.01 --> odd 1     urlname[0]
## 101.02 --> even 1   urlname[61]
## 102.01 --> even 0   urlname[72]
## 103.01 --> even 0   urlname[84]

result = pd.DataFrame(industryname, columns=['產業別'])

# From 96_1 to 101_1 
for i in range(len(urlname[:61])):
    new_df = pd.DataFrame()
    final_df = pd.DataFrame()
    print('This is: ', columnsname[i])
    print('Done!')

    url = urlname[i]   
    res = requests.get(url) 
    chineese_encoding = res.apparent_encoding
    res.encoding = chineese_encoding
    soup = BeautifulSoup(res.text,"html.parser") 
    
    for kkk in odd:
        data = soup.select('table')[kkk]
        df = pd.read_html(data.prettify())
        dfs = df[1]
        new_df = new_df.append(dfs)   
    
    new_df.columns = new_df.columns.get_level_values(1)
    final = new_df.loc[new_df["公司名稱"] == "合計"].reset_index(drop= True)
    final_df[columnsname[i]] = final['當月營收']
    result[columnsname[i]] = final_df[columnsname[i]]
    

##---------------------------------------------------------##
# From 101_2 to 101_12 
for i in np.arange(61, 72, 1):
    new_df = pd.DataFrame()
    final_df = pd.DataFrame()
    print('This is: ', columnsname[i])
    print('Done!')

    url = urlname[i]   
    res = requests.get(url) 
    chineese_encoding = res.apparent_encoding
    res.encoding = chineese_encoding
    soup = BeautifulSoup(res.text,"html.parser") 
    
    for kkk in even:
        data = soup.select('table')[kkk]
        df = pd.read_html(data.prettify())
        dfs = df[1]
        new_df = new_df.append(dfs)   
    
    new_df.columns = new_df.columns.get_level_values(1)
    final = new_df.loc[new_df["公司名稱"] == "合計"].reset_index(drop= True)
    final_df[columnsname[i]] = final['當月營收']
    result[columnsname[i]] = final_df[columnsname[i]]
    
    
##---------------------------------------------------------##
# From 102_1 to the end 
for i in np.arange(72, 185, 1):
    new_df = pd.DataFrame()
    final_df = pd.DataFrame()
    print('This is: ', columnsname[i])
    print('Done!')

    url = urlname[i]   
    res = requests.get(url) 
    chineese_encoding = res.apparent_encoding
    res.encoding = chineese_encoding
    soup = BeautifulSoup(res.text,"html.parser") 
    
    for kkk in even:
        data = soup.select('table')[kkk]
        df = pd.read_html(data.prettify())
        dfs = df[0]
        new_df = new_df.append(dfs)   
    
    new_df.columns = new_df.columns.get_level_values(1)
    final = new_df.loc[new_df["公司名稱"] == "合計"].reset_index(drop= True)
    final_df[columnsname[i]] = final['當月營收']
    result[columnsname[i]] = final_df[columnsname[i]]
```
